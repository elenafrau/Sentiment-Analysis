```python
## Eyes Detection
import cv2
path = r"C:\Users\umang\Desktop\Open CV"
face_xml = path+"\\face.xml"
eye_xml = path+"\\eye.xml"
face_img = path+"\\faces.jpg"
img = cv2.imread(face_img)
face_cascade = cv2.CascadeClassifier(face_xml)
eye_cascade = cv2.CascadeClassifier(eye_xml)
faces = face_cascade.detectMultiScale(img,1.2,5) 
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
    roi_face = img[y:y+h,x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_face,1.1,6)
    for ex,ey,ew,eh in eyes:
        cv2.rectangle(roi_face,(ex,ey),(ex+ew,ey+eh),(255,0,0),1)
cv2.imshow("Faces",img)
cv2.waitKey(0)  ### 0 means any key
cv2.destroyAllWindows()
```

```python
## Smiles Detection
import cv2
path = r"C:\Users\umang\Desktop\Open CV"
face_xml = path+"\\face.xml"
smile_xml = path+"\\smile.xml"
face_img = path+"\\smiles.jpg"
img = cv2.imread(face_img)
face_cascade = cv2.CascadeClassifier(face_xml)
smile_cascade = cv2.CascadeClassifier(smile_xml)
faces = face_cascade.detectMultiScale(img,1.2,5) 
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
    roi_face = img[y:y+h,x:x+w]
    smiles = smile_cascade.detectMultiScale(roi_face,1.8,10)
    for ex,ey,ew,eh in smiles:
        cv2.rectangle(roi_face,(ex,ey),(ex+ew,ey+eh),(255,0,0),1)
cv2.imshow("Smiles",img)
cv2.waitKey(0)  ### 0 means any key
cv2.destroyAllWindows()
```

## WebCam

```python
import cv2
path = r"C:\Users\umang\Desktop\Open CV"
face_xml = path+"\\face.xml"
face_cascade = cv2.CascadeClassifier(face_xml)
cam = cv2.VideoCapture(0)   ### It will open the default webcam(0)
while True:
    ret,frame = cam.read()   ### ret- True/False, frame - image which is captured
    frame = cv2.flip(frame,1  ### 1 - x axis flip,  0 - yaxis-flip
    faces = face_cascade.detectMultiScale(frame,1.2,5)
    for x,y,w,h in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),2)
    cv2.imshow("Webcam",frame)
    if cv2.waitKey(1) == 27:      ## 27 is ASCII value of esc
        break
cam.release()
cv2.destroyAllWindows()
```

### Save the video

```python
import cv2
path = r"C:\Users\umang\Desktop\Open CV"
face_xml = path+"\\face.xml"
filename = path+"\\video.avi"
### famous encoders - WMV2, WMV1, MJPG, DIVX, XVID, H264
codec = cv2.VideoWriter_fourcc('X','V','I','D')  ### fourcc - four character code
framerate = 30  ### 30 fps
resolution = (640,480)
output = cv2.VideoWriter(filename,codec,framerate,resolution)
face_cascade = cv2.CascadeClassifier(face_xml)
cam = cv2.VideoCapture(0)   ### It will open the default webcam(0)
while True:
    ret,frame = cam.read()   ### ret- True/False, frame - image which is captured
    frame = cv2.flip(frame,1)  ### 1 - x axis flip,  0 - yaxis-flip
    faces = face_cascade.detectMultiScale(frame,1.2,5)
    for x,y,w,h in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),2)
    output.write(frame)
    cv2.imshow("Webcam",frame)
    if cv2.waitKey(1) == 27:      ## 27 is ASCII value of esc
        break
output.release()
cam.release()
cv2.destroyAllWindows()
```

### Object Detection By Color

```python
import cv2
import numpy as np
```

```python
cam = cv2.VideoCapture(0)
while True:
    ret,frame = cam.read()
    #### hsv - hue saturation value
    hsv = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    #### red colour
    low = np.array([140,150,0])
    high = np.array([180,255,255])
    mask = cv2.inRange(hsv,low,high)
    output = cv2.bitwise_and(frame,frame,mask=mask)
    cv2.imshow("HSV Image",hsv)
    cv2.imshow("Mask Image",mask)
    cv2.imshow("Output",output)
    if cv2.waitKey(1) == 27:
        break
cam.release()
cv2.destroyAllWindows()
```
